## Fix log — why the model wasn't learning

**Symptom:** train & val loss frozen at `0.2000`, embeddings collapsing.

**Cause:** the batch-hard **triplet** loss collapsed every description to a single point on these low-signal, boilerplate-heavy descriptions. When all embeddings coincide, every pairwise distance is ~0, so `relu(pos - neg + margin) = margin` — that's why the loss equalled the margin (0.2) exactly.

**Fix (cells 6 & 7):** switched to **InfoNCE / Multiple-Negatives-Ranking** loss (the alternative named in the task description). Its constant-embedding solution is the *worst* solution (`loss ≈ ln(batch)`), not a plateau, so it can't get stuck there. Cell 7 also prints a live `dist` (mean pairwise distance of val embeddings) — watch it climb off ~0; if it stays near 0 the model is collapsing.


```python
# ============================================================
# 0. Data loading, combining, filtering & feasibility stats
# ============================================================
import pandas as pd
import numpy as np

# ---- Load & combine the two job-description datasets ----
df1 = pd.read_csv("job_title_des_1.csv", index_col=0)   # columns: Job Title, Job Description
df2 = pd.read_csv("job_title_des_2.csv")                 # columns: Job Title, Description

df1 = df1.rename(columns={"Job Description": "Description"})
combined = pd.concat([df1, df2], ignore_index=True)

# Basic cleaning: drop rows with missing title/description
combined = combined.dropna(subset=["Job Title", "Description"]).reset_index(drop=True)
combined["Job Title"] = combined["Job Title"].str.strip()

# ---- Remove job titles that occur fewer than 10 times ----
MIN_COUNT = 10
title_counts = combined["Job Title"].value_counts()
keep_titles = title_counts[title_counts >= MIN_COUNT].index
df = combined[combined["Job Title"].isin(keep_titles)].reset_index(drop=True)

# ---- Feasibility stats (.md: "Questions to Determine Feasibility") ----
# NOTE: dataset size is computed from the data, NOT the stale 2403 in the .md.
final_counts = df["Job Title"].value_counts()
print(f"Combined rows (raw):              {len(combined)}")
print(f"Rows after removing <{MIN_COUNT} titles:   {len(df)}")
print(f"Unique job titles kept:          {final_counts.size}")
print(f"Avg descriptions per title:      {final_counts.mean():.2f}")
print(f"Min / Max per title:             {final_counts.min()} / {final_counts.max()}")
print("\nClass imbalance across the kept titles (value counts):")
print(final_counts)
```

```python
# ============================================================
# 1. Tokenization  ->  integer ids  (Tokens -> [15, 832, ...])
# ============================================================
import re
from collections import Counter

def tokenize(text):
    return re.findall(r"[a-z]+", str(text).lower())

df["tokens"] = df["Description"].apply(tokenize)

# ---- Build vocabulary (index 0 reserved for <pad> -> padding_idx=0) ----
PAD, UNK = "<pad>", "<unk>"
MIN_TOKEN_FREQ = 2                     # drop very rare tokens (reduce overfitting)
counter = Counter(tok for toks in df["tokens"] for tok in toks)

itos = [PAD, UNK] + [w for w, c in counter.most_common() if c >= MIN_TOKEN_FREQ]
stoi = {w: i for i, w in enumerate(itos)}
vocab_size = len(itos)
PAD_IDX = stoi[PAD]                    # == 0

MAX_LEN = 100                          # avg length ~100 tokens (see .md "Training Time")
def encode(tokens):
    ids = [stoi.get(t, stoi[UNK]) for t in tokens][:MAX_LEN]
    return ids if ids else [stoi[UNK]]

df["input_ids"] = df["tokens"].apply(encode)
print(f"Vocabulary size: {vocab_size}  (PAD_IDX={PAD_IDX})")
```

```python
# ============================================================
# 2. Label encoding + stratified train/val split
# ============================================================
from sklearn.model_selection import train_test_split

# Encode job titles -> integer class labels (these define positives/negatives)
classes = sorted(df["Job Title"].unique())
class_to_idx = {c: i for i, c in enumerate(classes)}
df["label"] = df["Job Title"].map(class_to_idx)
num_classes = len(classes)

# Stratify so every title is represented in both splits despite imbalance.
train_df, val_df = train_test_split(
    df, test_size=0.15, random_state=42, stratify=df["label"]
)
train_df = train_df.reset_index(drop=True)
val_df = val_df.reset_index(drop=True)
print(f"Train: {len(train_df)}  |  Val: {len(val_df)}  |  Classes: {num_classes}")
```

```python
# ============================================================
# 3. Pretrained FastText embedding matrix  (.md step 1)
# ============================================================
# 300-d FastText vectors, aligned to our vocab. Downloads once via gensim.
import torch
import gensim.downloader as api

EMB_DIM = 300
print("Loading FastText vectors (first run downloads ~1 GB)...")
ft = api.load("fasttext-wiki-news-subwords-300")

rng = np.random.default_rng(42)
embedding_matrix = rng.normal(scale=0.1, size=(vocab_size, EMB_DIM)).astype("float32")
embedding_matrix[PAD_IDX] = 0.0                      # pad row stays zero

hits = 0
for word, idx in stoi.items():
    if word in ft:
        embedding_matrix[idx] = ft[word]
        hits += 1
embedding_matrix = torch.tensor(embedding_matrix)
print(f"Pretrained vectors matched: {hits}/{vocab_size} tokens")
```

```python
# ============================================================
# 4. Dataset, padding collate & BALANCED batch sampler
#    -> this is how we handle the job-title imbalance (TRAIN only)
# ============================================================
from torch.utils.data import Dataset, DataLoader, Sampler

class JobDataset(Dataset):
    def __init__(self, frame):
        self.ids = frame["input_ids"].tolist()
        self.labels = frame["label"].tolist()
    def __len__(self):
        return len(self.ids)
    def __getitem__(self, i):
        return self.ids[i], self.labels[i]

def collate(batch):
    seqs, labels = zip(*batch)
    lengths = torch.tensor([len(s) for s in seqs])
    maxlen = int(lengths.max())
    padded = torch.full((len(seqs), maxlen), PAD_IDX, dtype=torch.long)
    for i, s in enumerate(seqs):
        padded[i, :len(s)] = torch.tensor(s, dtype=torch.long)
    return padded, lengths, torch.tensor(labels, dtype=torch.long)

class PKSampler(Sampler):
    """Balanced P x K batch sampler: each batch = P titles x K descriptions.
    Sampling equal K per title equalizes frequent vs rare titles (imbalance)
    and guarantees >=K positives per class, which metric-learning losses
    (triplet / InfoNCE) need. TRAIN ONLY -- never use this for validation."""
    def __init__(self, labels, P=8, K=4, seed=42):
        self.labels = np.array(labels)
        self.class_to_indices = {c: np.where(self.labels == c)[0]
                                 for c in np.unique(self.labels)}
        self.classes = list(self.class_to_indices.keys())
        self.P = min(P, len(self.classes))     # guard: can't pick more classes than exist
        self.K = K
        self.g = np.random.default_rng(seed)
        self.num_batches = max(1, len(self.labels) // (self.P * self.K))
    def __iter__(self):
        for _ in range(self.num_batches):
            chosen = self.g.choice(self.classes, size=self.P, replace=False)
            batch = []
            for c in chosen:
                idxs = self.class_to_indices[c]
                pick = self.g.choice(idxs, size=self.K, replace=len(idxs) < self.K)
                batch.extend(pick.tolist())
            yield batch
    def __len__(self):
        return self.num_batches

P, K = 8, 4                                    # 8 titles x 4 desc = batch_size 32 (.md)
train_loader = DataLoader(
    JobDataset(train_df), collate_fn=collate,
    batch_sampler=PKSampler(train_df["label"].tolist(), P, K, seed=42))

# VALIDATION FIX: the PK sampler must NEVER be used for validation.
# After the split many val titles hold only 1-3 samples, so PK's `replace=True`
# draw duplicates the SAME description into a batch as its own positive
# (sim(A, A) ~= 1) -> an artificially tiny, noisy loss that measures duplicates,
# not generalization. It is also stochastic (fresh random batches every epoch)
# and gives a varying #positives per anchor, which changes the InfoNCE scale
# batch-to-batch. Use a plain, deterministic, full-coverage loader instead.
val_loader = DataLoader(
    JobDataset(val_df), collate_fn=collate,
    batch_size=64, shuffle=False)
print(f"Train batches: {len(train_loader)}  |  Val batches: {len(val_loader)}")
```

```python
# ============================================================
# 5. The model:  FastText -> BiLSTM -> mean pool -> proj -> L2
# ============================================================
import torch.nn as nn
import torch.nn.functional as F

class JobEmbeddingModel(nn.Module):
    def __init__(self, embedding_matrix, hidden_size=128, num_layers=2,
                 proj_hidden=256, out_dim=128, dropout=0.3):
        super().__init__()
        n_vocab, emb_dim = embedding_matrix.shape

        # 1. Embedding layer (pretrained FastText, fine-tunable)
        self.embedding = nn.Embedding(n_vocab, emb_dim, padding_idx=PAD_IDX)
        self.embedding.weight.data.copy_(embedding_matrix)
        self.embedding.weight.requires_grad = True

        # 2. BiLSTM encoder (.md "Dataset Size Concern": hidden_size=128 for small data)
        self.lstm = nn.LSTM(
            input_size=emb_dim, hidden_size=hidden_size, num_layers=num_layers,
            batch_first=True, bidirectional=True, dropout=dropout)
        lstm_out = hidden_size * 2                 # 128 fwd + 128 bwd = 256

        # 4. Projection head
        self.proj = nn.Sequential(
            nn.Linear(lstm_out, proj_hidden),      # 256 -> 256
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(proj_hidden, out_dim),       # 256 -> 128
        )

    def forward(self, x, lengths=None):
        mask = (x != PAD_IDX).unsqueeze(-1).float()   # (B, T, 1)
        emb = self.embedding(x)                        # (B, T, 300)
        outputs, _ = self.lstm(emb)                    # (B, T, 256)

        # 3. Mean pooling over real tokens (padding-aware mean)
        summed = (outputs * mask).sum(dim=1)           # (B, 256)
        pooled = summed / mask.sum(dim=1).clamp(min=1e-9)

        z = self.proj(pooled)                          # (B, 128)
        # 5. L2 normalize -> final embedding
        return F.normalize(z, p=2, dim=1)
```

```python
# ============================================================
# 6. Metric-learning loss
#    FIX: batch-hard TRIPLET  ->  InfoNCE / Multiple-Negatives-Ranking
# ------------------------------------------------------------
# WHY THE SWAP: on these boilerplate-heavy, low-signal job descriptions the
# batch-hard triplet loss collapsed every embedding to a single point. At
# collapse all pairwise distances are ~0, so relu(pos - neg + margin) = margin
# -- which is exactly why train/val froze at 0.2000 (== the old margin).
# InfoNCE's trivial constant solution is instead the WORST solution
# (loss ~= ln(batch)), not a flat plateau, so the optimizer is pushed away from
# collapse. Because each PK batch holds K descriptions per title, every anchor
# has several positives -> this is the supervised-contrastive form of InfoNCE,
# a.k.a. Multiple Negatives Ranking Loss (the alternative named in the .md).
def info_nce_loss(embeddings, labels, temp=0.07):
    # embeddings are already L2-normalized -> dot product == cosine similarity
    sim = embeddings @ embeddings.t() / temp                 # (B, B)
    n = sim.size(0)
    eye = torch.eye(n, dtype=torch.bool, device=sim.device)
    sim = sim.masked_fill(eye, float("-inf"))                # drop self-similarity
    labels = labels.unsqueeze(1)
    pos_mask = (labels == labels.t()) & ~eye                 # same title = positive
    # log-softmax over all other items, then average log-prob of the positives
    logp = sim - torch.logsumexp(sim, dim=1, keepdim=True)
    logp = torch.where(pos_mask, logp, torch.zeros_like(logp))
    loss = -logp.sum(dim=1) / pos_mask.sum(dim=1).clamp(min=1)
    return loss.mean()


# --- OLD loss, kept ONLY for reference. This is what caused the collapse and is
#     no longer called. Left in so you can compare the two approaches. ---
def batch_hard_triplet_loss(embeddings, labels, margin=0.2):
    dist = torch.cdist(embeddings, embeddings, p=2)          # (B, B)
    labels = labels.unsqueeze(1)
    same = labels == labels.t()
    eye = torch.eye(len(labels), dtype=torch.bool, device=embeddings.device)
    pos_mask = same & ~eye
    neg_mask = ~same
    hardest_pos = dist.masked_fill(~pos_mask, float("-inf")).max(dim=1).values
    hardest_neg = dist.masked_fill(~neg_mask, float("inf")).min(dim=1).values
    valid = torch.isfinite(hardest_pos) & torch.isfinite(hardest_neg)
    if not valid.any():
        return embeddings.sum() * 0.0
    loss = F.relu(hardest_pos[valid] - hardest_neg[valid] + margin)
    return loss.mean()
```

```python
# ============================================================
# 7. Training  (.md "Overfitting" hyperparameters)
#    Monitors:
#      - RETRIEVAL metrics R@1 / R@5 / MRR  <- the real signal, checkpoint on it
#      - val InfoNCE loss (now deterministic) for reference only
#      - collapse monitor 'dist' (mean pairwise dist of val embeddings)
#    In metric learning it is normal for train loss down / val loss UP while
#    retrieval keeps improving -- so we select the checkpoint on MRR, not loss.
# ============================================================
import time
from tqdm.auto import tqdm

device = torch.device(
    "mps" if torch.backends.mps.is_available()
    else "cuda" if torch.cuda.is_available() else "cpu")

model = JobEmbeddingModel(embedding_matrix, hidden_size=128, dropout=0.3).to(device)

# Hyperparameters. TEMP is the InfoNCE temperature (replaces the triplet MARGIN).
LR, WEIGHT_DECAY = 1e-3, 1e-5
EPOCHS, TEMP = 50, 0.07

optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)

def run_epoch(loader, train, epoch):
    model.train() if train else model.eval()
    total, n = 0.0, 0
    stage = "train" if train else " val "
    # Per-batch progress bar (leave=False -> it clears after each epoch).
    inner = tqdm(loader, desc=f"Epoch {epoch}/{EPOCHS} [{stage}]",
                 unit="batch", leave=False)
    with torch.set_grad_enabled(train):
        for x, lengths, labels in inner:
            x, labels = x.to(device), labels.to(device)
            z = model(x, lengths)
            loss = info_nce_loss(z, labels, temp=TEMP)
            if train:
                optimizer.zero_grad()
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 5.0)
                optimizer.step()
            total += loss.item(); n += 1
            inner.set_postfix(loss=f"{loss.item():.4f}")
    inner.close()
    return total / max(n, 1)

@torch.no_grad()
def embed_frame(frame, batch=64):
    """Embed every row of a dataframe -> (N, 128) embeddings + (N,) labels."""
    model.eval()
    ids = frame["input_ids"].tolist()
    out = []
    for i in range(0, len(ids), batch):
        chunk = ids[i:i + batch]
        lengths = torch.tensor([len(s) for s in chunk])
        maxlen = int(lengths.max())
        padded = torch.full((len(chunk), maxlen), PAD_IDX, dtype=torch.long)
        for j, s in enumerate(chunk):
            padded[j, :len(s)] = torch.tensor(s, dtype=torch.long)
        out.append(model(padded.to(device), lengths).cpu())
    return torch.cat(out, dim=0), torch.tensor(frame["label"].tolist())

@torch.no_grad()
def retrieval_metrics():
    """Task-aligned eval (description -> nearest job title). Build title
    centroids from TRAIN embeddings, then for each VAL description measure the
    rank of its true title. Returns Recall@1, Recall@5, MRR -- far more
    informative than the InfoNCE val loss for an embedding/retrieval model."""
    model.eval()
    tr_emb, tr_lab = embed_frame(train_df)
    va_emb, va_lab = embed_frame(val_df)
    cls = torch.unique(tr_lab)
    cents = torch.stack([F.normalize(tr_emb[tr_lab == c].mean(0), dim=0) for c in cls])
    sims = va_emb @ cents.t()                       # (Nval, num_classes) cosine
    ranked = cls[sims.argsort(dim=1, descending=True)]
    true = va_lab.unsqueeze(1)
    hits = ranked == true
    r1 = hits[:, 0].float().mean().item()
    r5 = hits[:, :min(5, hits.size(1))].any(dim=1).float().mean().item()
    mrr = (1.0 / (hits.float().argmax(dim=1) + 1).float()).mean().item()
    return r1, r5, mrr

@torch.no_grad()
def mean_pairwise_dist(loader, max_batches=4):
    """Collapse detector: average L2 distance between val embeddings.
    A value near 0 means every description maps to the same point (collapse)."""
    model.eval()
    zs = []
    for bi, (x, lengths, labels) in enumerate(loader):
        zs.append(model(x.to(device), lengths))
        if bi + 1 >= max_batches:
            break
    z = torch.cat(zs, dim=0)
    return torch.cdist(z, z).mean().item()

# Epoch-level bar: tqdm's own [elapsed<remaining], an explicit ETA, retrieval
# metrics, and the collapse monitor 'dist'.
best_mrr, start = -1.0, time.time()
pbar = tqdm(range(1, EPOCHS + 1), desc="Training", unit="epoch")
for epoch in pbar:
    tr = run_epoch(train_loader, True, epoch)
    va = run_epoch(val_loader, False, epoch)       # deterministic val loss (ref only)
    r1, r5, mrr = retrieval_metrics()              # <-- the real generalization signal
    dist = mean_pairwise_dist(val_loader)          # <-- collapse monitor

    # ---- ETA: avg epoch time x remaining epochs ----
    avg_epoch = (time.time() - start) / epoch
    eta_min = avg_epoch * (EPOCHS - epoch) / 60
    pbar.set_postfix(val=f"{va:.3f}", R1=f"{r1:.2f}", R5=f"{r5:.2f}",
                     MRR=f"{mrr:.3f}", dist=f"{dist:.2f}", eta=f"~{eta_min:.1f}m")
    if dist < 0.05:
        tqdm.write(f"  [warn] epoch {epoch}: mean pairwise dist {dist:.4f} "
                   f"-> embeddings look collapsed, not learning")

    # Checkpoint on RETRIEVAL quality (MRR), NOT val loss. Training always runs
    # the full EPOCHS (no early stopping) -- we just keep the best snapshot.
    if mrr > best_mrr + 1e-4:
        best_mrr = mrr
        torch.save(model.state_dict(), "job_embedding_best.pt")

print(f"Training complete ({EPOCHS} epochs). Best val MRR: {best_mrr:.4f}")
```

```python
# ============================================================
# 8. Compute & save per-title embeddings  (consumed by app.py)
# ============================================================
import json, os

os.makedirs("artifacts", exist_ok=True)

# Reload the best checkpoint (lowest val loss seen during training)
model.load_state_dict(torch.load("job_embedding_best.pt", map_location=device))
model.eval()

@torch.no_grad()
def embed_descriptions(id_lists, batch=64):
    """Embed a list of token-id lists -> (N, 128) L2-normalized tensor."""
    out = []
    for i in range(0, len(id_lists), batch):
        chunk = id_lists[i:i + batch]
        lengths = torch.tensor([len(s) for s in chunk])
        maxlen = int(lengths.max())
        padded = torch.full((len(chunk), maxlen), PAD_IDX, dtype=torch.long)
        for j, s in enumerate(chunk):
            padded[j, :len(s)] = torch.tensor(s, dtype=torch.long)
        out.append(model(padded.to(device), lengths).cpu())
    return torch.cat(out, dim=0)

# Embed every description, then average per title -> title centroid embedding
desc_emb = embed_descriptions(df["input_ids"].tolist())          # (N, 128)

title_vecs, title_names, title_counts = [], [], []
for label, pos in df.groupby("label").groups.items():
    centroid = desc_emb[list(pos)].mean(dim=0)                   # mean of unit vectors
    centroid = F.normalize(centroid, p=2, dim=0)                 # re-normalize -> unit
    title_vecs.append(centroid)
    title_names.append(classes[label])
    title_counts.append(len(pos))

title_embeddings = torch.stack(title_vecs)                       # (num_titles, 128)

# ---- Save the title embeddings + everything app.py needs to embed a query ----
torch.save({
    "title_embeddings": title_embeddings,          # (num_titles, 128), unit vectors
    "titles": title_names,                         # aligned with rows above
    "counts": title_counts,                        # #descriptions per title
}, "artifacts/title_embeddings.pt")

with open("artifacts/vocab.json", "w") as f:
    json.dump({
        "stoi": stoi,
        "unk_idx": stoi[UNK],
        "pad_idx": PAD_IDX,
        "max_len": MAX_LEN,
        "emb_dim": EMB_DIM,
        "hidden_size": 128,
        "num_layers": 2,
        "proj_hidden": 256,
        "out_dim": 128,
        "vocab_size": vocab_size,
    }, f)

print(f"Saved {title_embeddings.shape[0]} title embeddings -> artifacts/title_embeddings.pt")
print("Saved tokenizer/config           -> artifacts/vocab.json")
print("Model weights already at          job_embedding_best.pt")
```

```python
# ============================================================
# 9. Visualize the learned embedding space (UMAP)
# ============================================================
# 2-D UMAP of the per-description embeddings (desc_emb, computed in cell 8),
# coloured by job title, with the per-title centroids overlaid as stars. Tight,
# well-separated colour clusters => the encoder learned title-discriminative
# representations.
try:
    import umap
except ImportError:
    import subprocess, sys
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "umap-learn"],
                   check=True)
    import umap

import matplotlib.pyplot as plt

# desc_emb (N,128) and title_embeddings (T,128) are already L2-normalized, so we
# fit UMAP on both together (cosine metric) to keep points + centroids aligned.
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, metric="cosine", random_state=42)
emb_all = torch.cat([desc_emb, title_embeddings], dim=0).numpy()
coords = reducer.fit_transform(emb_all)
desc_xy = coords[:len(desc_emb)]
title_xy = coords[len(desc_emb):]

labels = df["label"].to_numpy()
cmap = plt.get_cmap("tab20", num_classes)

fig, ax = plt.subplots(figsize=(13, 10))
for c in range(num_classes):
    m = labels == c
    ax.scatter(desc_xy[m, 0], desc_xy[m, 1], s=8, alpha=0.5,
               color=cmap(c), label=classes[c])

# overlay per-title centroids (black-edged stars) + text labels
ax.scatter(title_xy[:, 0], title_xy[:, 1], s=200, marker="*",
           c=[cmap(i) for i in range(num_classes)],
           edgecolors="black", linewidths=1.2, zorder=5)
for i, name in enumerate(title_names):
    ax.annotate(name, (title_xy[i, 0], title_xy[i, 1]), fontsize=8, weight="bold",
                xytext=(3, 3), textcoords="offset points", zorder=6)

ax.set_title("UMAP of job-description embeddings "
             "(colour = title, ★ = title centroid)")
ax.set_xlabel("UMAP-1"); ax.set_ylabel("UMAP-2")
ax.legend(markerscale=2, fontsize=7, ncol=2, loc="best", framealpha=0.9)
plt.tight_layout()
plt.savefig("artifacts/umap_embeddings.png", dpi=130)
plt.show()
print("Saved UMAP figure -> artifacts/umap_embeddings.png")
```
