# Fashion Similarity Search Engine — Offline Pipeline

End-to-end **content-based image retrieval** for fashion products, built per
`fashion_similarity_search_engine.md`.

This notebook is the **offline stage**. Run it top-to-bottom to produce every
artifact the Streamlit app (`app_backup.py`) consumes:

```
Dataset -> Fine-Tune ResNet50 -> Save Best Model -> Generate Embeddings
        -> Build FAISS Index -> Generate UMAP Coords -> Save Artifacts
```

**Data layout expected** (Kaggle *fashion-product-images-small*):

```
data/
├── images/            # 15970.jpg, 15971.jpg, ...
└── styles.csv         # gender, masterCategory, subCategory, articleType, ...
```

**Extra deps** (not in the base venv): `torchvision`, `faiss-cpu`, `umap-learn`,
`matplotlib`.  Install with:

```
pip install torchvision faiss-cpu umap-learn matplotlib
```

**Artifacts produced** in `artifacts/`: `best_model.pt`, `image_embeddings.npy`,
`image_paths.pkl`, `metadata.pkl`, `faiss.index`, `umap_coords.npy`,
`label_encoder.pkl`, `training_history.json`.

````python
# ============================================================
# 0. Configuration
#    Change TARGET_COL to switch the classification target with NO other edits.
# ============================================================
from pathlib import Path

DATA_DIR   = Path("data")
IMAGES_DIR = DATA_DIR / "images"
STYLES_CSV = DATA_DIR / "styles.csv"
ARTIFACTS  = Path("artifacts")
ARTIFACTS.mkdir(parents=True, exist_ok=True)

# Classification target: "articleType" (preferred) | "subCategory" | "masterCategory"
TARGET_COL = "articleType"

MIN_CLASS_COUNT = 50      # drop ultra-rare classes so stratified split is stable
IMG_SIZE        = 224
BATCH_SIZE      = 64
VAL_SPLIT       = 0.15
SEED            = 42

# Two-phase fine-tuning schedule
HEAD_EPOCHS     = 3       # phase 1: train classifier head only (backbone frozen)
FINETUNE_EPOCHS = 5       # phase 2: unfreeze layer4 + head, fine-tune end-to-end

BACKBONE_LR   = 1e-5      # small LR for pretrained backbone
CLASSIFIER_LR = 1e-4      # larger LR for the fresh head
WEIGHT_DECAY  = 1e-4

EMB_DIM = 2048            # ResNet50 global-average-pool feature size
TOP_K   = 12             # default neighbours for retrieval

print(f"Target column : {TARGET_COL}")
print(f"Artifacts dir : {ARTIFACTS.resolve()}")
````

````python
# ============================================================
# 1. Imports & device
# ============================================================
import json, pickle, time
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

from torchvision import transforms
from torchvision.models import resnet50, ResNet50_Weights

from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

torch.manual_seed(SEED)
np.random.seed(SEED)

device = torch.device(
    "cuda" if torch.cuda.is_available()
    else "mps" if torch.backends.mps.is_available() else "cpu")
print("Device:", device)
````

````python
# ============================================================
# 2. Load & clean metadata (styles.csv)
# ============================================================
# styles.csv has a few malformed rows (stray commas in productDisplayName),
# so skip bad lines. Keep only rows whose image file actually exists on disk.
styles = pd.read_csv(STYLES_CSV, on_bad_lines="skip")
styles["id"] = styles["id"].astype(str)
styles["image"] = styles["id"] + ".jpg"

styles = styles.dropna(subset=[TARGET_COL])
styles = styles[styles["image"].apply(lambda f: (IMAGES_DIR / f).exists())]

# drop ultra-rare classes so the stratified split is well-defined
counts = styles[TARGET_COL].value_counts()
keep = counts[counts >= MIN_CLASS_COUNT].index
styles = styles[styles[TARGET_COL].isin(keep)].reset_index(drop=True)

print(f"Usable images : {len(styles)}")
print(f"Classes kept  : {styles[TARGET_COL].nunique()} (>= {MIN_CLASS_COUNT} each)")
print(styles[TARGET_COL].value_counts().head(10))
````

````python
# ============================================================
# 3. Label-encode target + stratified train/val split
# ============================================================
label_encoder = LabelEncoder()
styles["label"] = label_encoder.fit_transform(styles[TARGET_COL])
num_classes = len(label_encoder.classes_)

train_df, val_df = train_test_split(
    styles, test_size=VAL_SPLIT, random_state=SEED, stratify=styles["label"])
train_df = train_df.reset_index(drop=True)
val_df   = val_df.reset_index(drop=True)
print(f"Train: {len(train_df)}  |  Val: {len(val_df)}  |  Classes: {num_classes}")
````

````python
# ============================================================
# 4. Dataset & image transforms (augmentation on train, plain on eval)
# ============================================================
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]

train_tf = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(0.1, 0.1, 0.1),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])
eval_tf = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

class FashionDataset(Dataset):
    def __init__(self, frame, transform):
        self.files  = frame["image"].tolist()
        self.labels = frame["label"].tolist()
        self.transform = transform
    def __len__(self):
        return len(self.files)
    def __getitem__(self, i):
        img = Image.open(IMAGES_DIR / self.files[i]).convert("RGB")
        return self.transform(img), self.labels[i]

train_loader = DataLoader(FashionDataset(train_df, train_tf),
                          batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
val_loader   = DataLoader(FashionDataset(val_df, eval_tf),
                          batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
````

````python
# ============================================================
# 5. Model: pretrained ResNet50 + new classification head
#    Input -> Resize(224) -> Normalize -> ResNet50 -> GAP -> FC -> Softmax
# ============================================================
def build_model(num_classes):
    model = resnet50(weights=ResNet50_Weights.IMAGENET1K_V2)
    in_features = model.fc.in_features            # 2048 after global-average-pool
    model.fc = nn.Linear(in_features, num_classes)
    return model

model = build_model(num_classes).to(device)

def set_trainable(model, train_layer4, train_head=True):
    """Phase control: freeze all, then optionally unfreeze layer4 and/or head."""
    for p in model.parameters():
        p.requires_grad = False
    for p in model.fc.parameters():
        p.requires_grad = train_head
    if train_layer4:
        for p in model.layer4.parameters():
            p.requires_grad = True

set_trainable(model, train_layer4=False)          # phase 1: head only
print("Trainable params (phase 1):",
      sum(p.numel() for p in model.parameters() if p.requires_grad))
````

````python
# ============================================================
# 6. Train / evaluate helpers (loss + top-1 / top-5 accuracy)
# ============================================================
from tqdm.auto import tqdm

criterion = nn.CrossEntropyLoss()

@torch.no_grad()
def topk_correct(logits, targets, ks=(1, 5)):
    maxk = min(max(ks), logits.size(1))
    _, pred = logits.topk(maxk, dim=1)
    correct = pred.t().eq(targets.view(1, -1))
    return {k: correct[:min(k, maxk)].any(dim=0).float().sum().item() for k in ks}

def run_epoch(loader, optimizer=None, desc="train"):
    train = optimizer is not None
    model.train() if train else model.eval()
    tot_loss, n, c1, c5 = 0.0, 0, 0.0, 0.0
    bar = tqdm(loader, desc=desc, leave=False, unit="batch")
    with torch.set_grad_enabled(train):
        for x, y in bar:
            x, y = x.to(device), y.to(device)
            logits = model(x)
            loss = criterion(logits, y)
            if train:
                optimizer.zero_grad(); loss.backward(); optimizer.step()
            bs = y.size(0); tot_loss += loss.item() * bs; n += bs
            ck = topk_correct(logits, y); c1 += ck[1]; c5 += ck[5]
            bar.set_postfix(loss=f"{loss.item():.3f}")
    bar.close()
    return tot_loss / n, c1 / n, c5 / n
````

````python
# ============================================================
# 7. Two-phase fine-tuning
#    Phase 1: classifier head only (backbone frozen)
#    Phase 2: unfreeze layer4 + head -> fine-tune end-to-end (discriminative LRs)
#    Checkpoint best by val top-1; log training_history.json
# ============================================================
history = {"train_loss": [], "val_loss": [],
           "val_top1": [], "val_top5": [], "phase": []}
best_top1 = -1.0
start = time.time()

def optimizer_for_phase(phase):
    if phase == 1:                                   # only the head trains
        return torch.optim.AdamW(model.fc.parameters(),
                                 lr=CLASSIFIER_LR, weight_decay=WEIGHT_DECAY)
    return torch.optim.AdamW([                        # phase 2: split LRs
        {"params": model.layer4.parameters(), "lr": BACKBONE_LR},
        {"params": model.fc.parameters(),     "lr": CLASSIFIER_LR},
    ], weight_decay=WEIGHT_DECAY)

def train_phase(phase, epochs):
    global best_top1
    opt = optimizer_for_phase(phase)
    for ep in range(1, epochs + 1):
        tr_loss, _, _   = run_epoch(train_loader, opt, desc=f"P{phase} ep{ep} train")
        va_loss, t1, t5 = run_epoch(val_loader,   None, desc=f"P{phase} ep{ep} val")
        history["train_loss"].append(tr_loss); history["val_loss"].append(va_loss)
        history["val_top1"].append(t1); history["val_top5"].append(t5)
        history["phase"].append(phase)
        print(f"[P{phase} {ep}/{epochs}] train {tr_loss:.3f} | "
              f"val {va_loss:.3f} | top1 {t1:.3f} | top5 {t5:.3f}")
        if t1 > best_top1:
            best_top1 = t1
            torch.save({"state_dict": model.state_dict(),
                        "num_classes": num_classes,
                        "target_col": TARGET_COL}, ARTIFACTS / "best_model.pt")

set_trainable(model, train_layer4=False); train_phase(1, HEAD_EPOCHS)   # phase 1
set_trainable(model, train_layer4=True);  train_phase(2, FINETUNE_EPOCHS)  # phase 2

with open(ARTIFACTS / "training_history.json", "w") as f:
    json.dump(history, f, indent=2)
print(f"\nDone in {(time.time()-start)/60:.1f} min. Best val top-1: {best_top1:.3f}")
````

````python
# ============================================================
# 7b. Training curves + confusion matrix (best checkpoint)
# ============================================================
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].plot(history["train_loss"], label="train"); ax[0].plot(history["val_loss"], label="val")
ax[0].set_title("Loss"); ax[0].set_xlabel("epoch"); ax[0].legend()
ax[1].plot(history["val_top1"], label="top-1"); ax[1].plot(history["val_top5"], label="top-5")
ax[1].set_title("Val accuracy"); ax[1].set_xlabel("epoch"); ax[1].legend()
plt.tight_layout(); plt.savefig(ARTIFACTS / "training_curves.png", dpi=120); plt.show()

ckpt = torch.load(ARTIFACTS / "best_model.pt", map_location=device)
model.load_state_dict(ckpt["state_dict"]); model.eval()

y_true, y_pred = [], []
with torch.no_grad():
    for x, y in tqdm(val_loader, desc="confusion", leave=False):
        y_pred.extend(model(x.to(device)).argmax(1).cpu().tolist())
        y_true.extend(y.tolist())
cm = confusion_matrix(y_true, y_pred)
fig, ax = plt.subplots(figsize=(8, 7))
im = ax.imshow(cm, cmap="viridis"); fig.colorbar(im)
ax.set_title(f"Confusion matrix ({num_classes} classes)")
ax.set_xlabel("predicted"); ax.set_ylabel("true")
plt.tight_layout(); plt.savefig(ARTIFACTS / "confusion_matrix.png", dpi=120); plt.show()
````

````python
# ============================================================
# 8. Feature extraction: fine-tuned backbone -> 2048-d L2-normalized embeddings
# ============================================================
# Reload best checkpoint, then swap the classification head for Identity so the
# network outputs the 2048-d global-average-pooled feature vector.
ckpt = torch.load(ARTIFACTS / "best_model.pt", map_location=device)
encoder = build_model(ckpt["num_classes"])
encoder.load_state_dict(ckpt["state_dict"])
encoder.fc = nn.Identity()                         # drop classifier -> feature extractor
encoder = encoder.to(device).eval()

@torch.no_grad()
def extract_embeddings(frame, batch=BATCH_SIZE):
    loader = DataLoader(FashionDataset(frame, eval_tf),
                        batch_size=batch, shuffle=False, num_workers=2)
    feats = []
    for x, _ in tqdm(loader, desc="embedding", unit="batch"):
        z = encoder(x.to(device))                  # (B, 2048)
        z = F.normalize(z, p=2, dim=1)             # L2 -> inner product == cosine
        feats.append(z.cpu().numpy())
    return np.concatenate(feats, axis=0).astype("float32")

# Embed EVERY usable image (keep metadata order so all artifacts stay aligned)
all_df = styles.reset_index(drop=True)
embeddings = extract_embeddings(all_df)
print("Embeddings:", embeddings.shape)
````

````python
# ============================================================
# 9. Save embeddings + image paths + metadata + label encoder
#    (row order is identical across all four -> the app indexes by row i)
# ============================================================
np.save(ARTIFACTS / "image_embeddings.npy", embeddings)

image_paths = [str(IMAGES_DIR / f) for f in all_df["image"]]
with open(ARTIFACTS / "image_paths.pkl", "wb") as f:
    pickle.dump(image_paths, f)

meta_cols = [c for c in ["id", "productDisplayName", "gender", "masterCategory",
                         "subCategory", "articleType", "baseColour", "season",
                         "usage", "image"] if c in all_df.columns]
with open(ARTIFACTS / "metadata.pkl", "wb") as f:
    pickle.dump(all_df[meta_cols].reset_index(drop=True), f)

with open(ARTIFACTS / "label_encoder.pkl", "wb") as f:
    pickle.dump(label_encoder, f)

print("Saved image_embeddings.npy / image_paths.pkl / metadata.pkl / label_encoder.pkl")
````

````python
# ============================================================
# 10. Build FAISS index (IndexFlatIP == cosine on unit-normalized vectors)
# ============================================================
import faiss

index = faiss.IndexFlatIP(EMB_DIM)                 # inner product on unit vectors
index.add(embeddings)
faiss.write_index(index, str(ARTIFACTS / "faiss.index"))
print(f"FAISS index built with {index.ntotal} vectors -> artifacts/faiss.index")
````

````python
# ============================================================
# 11. UMAP: 2-D projection of the embedding space (for the Explorer page)
# ============================================================
import umap

reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, metric="cosine", random_state=SEED)
umap_coords = reducer.fit_transform(embeddings).astype("float32")
np.save(ARTIFACTS / "umap_coords.npy", umap_coords)
print("UMAP coords:", umap_coords.shape, "-> artifacts/umap_coords.npy")
````

````python
# ============================================================
# 12. Sanity check: retrieve nearest neighbours for one image
# ============================================================
q = 0
D, I = index.search(embeddings[q:q + 1], TOP_K)
print("Query:", all_df.loc[q, "productDisplayName"])
for rank, (idx, score) in enumerate(zip(I[0], D[0]), 1):
    print(f"  {rank:2d}. sim={score:.3f}  {all_df.loc[idx, 'productDisplayName']} "
          f"({all_df.loc[idx, TARGET_COL]})")
````

````python
# ============================================================
# 13. Artifact summary
# ============================================================
for p in sorted(ARTIFACTS.glob("*")):
    print(f"  {p.name:26s} {p.stat().st_size / 1e6:8.2f} MB")
print("\nOffline pipeline complete. Launch the app with:")
print("  streamlit run app_backup.py")
````
