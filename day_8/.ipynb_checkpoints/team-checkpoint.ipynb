## Country Data — помощь странам

**Легенда:** мы консультанты гуманитарного фонда. У нас 167 стран с 9 социально-экономическими и медицинскими показателями, но **нет целевого столбца**. Нужно самим найти группы стран по уровню развития, превратить их в метку и проверить, можно ли эту метку потом предсказывать.

**Цепочка работы:**
1. EDA и подготовка (масштабирование)
2. Кластеризация (KMeans) — выбор k по локтю и силуэту
3. Создание прогнозного столбца — имена кластерам по смыслу
4. PCA для проверки — проекция на плоскость
5. Классификация — предсказываем нашу метку, честные метрики


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, classification_report)

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
RANDOM_STATE = 42
```

## Шаг 1. EDA и подготовка

Смотрим на данные раньше, чем на модели: размер, масштабы, пропуски, выбросы, корреляции. Затем приводим к одному масштабу `StandardScaler` — иначе `income`/`gdpp` (тысячи) перетянут кластеризацию и PCA на себя.

```python
df = pd.read_csv("Country-data.csv")
print("Размер:", df.shape)
df.head()
```

```python
# Пропуски и типы
print("Пропусков всего:", df.isna().sum().sum())
df.info()
```

```python
# Описательная статистика — видно, что масштабы признаков очень разные
df.describe().T
```

```python
# country — это идентификатор, в модель не идёт. Числовые признаки:
features = ["child_mort", "exports", "health", "imports", "income",
            "inflation", "life_expec", "total_fer", "gdpp"]
X = df[features].copy()
X.shape
```

```python
# Распределения признаков
X.hist(bins=30, figsize=(12, 9))
plt.suptitle("Распределения признаков", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Матрица корреляций
plt.figure(figsize=(9, 7))
sns.heatmap(X.corr(), annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Матрица корреляций")
plt.show()
```

**Находка EDA (которая нас удивила):** `child_mort` (детская смертность), `total_fer` (рождаемость) и `life_expec` (продолжительность жизни) образуют очень сильный блок корреляций, а `income`, `gdpp` и `life_expec` тянутся вместе. То есть «бедность», «высокая детская смертность» и «высокая рождаемость» — это фактически одна и та же ось. Это подсказывает, что осмысленные кластеры будут именно по **уровню развития**, и что двух главных компонент PCA, скорее всего, хватит.

```python
# Масштабирование — обязательный шаг перед KMeans и PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=features)
X_scaled.describe().T[["mean", "std", "min", "max"]].round(2)
```

## Шаг 2. Кластеризация (KMeans)

Меток нет — структуру ищем сами. Число кластеров `k` не угадываем: строим **метод локтя** (inertia от k) и **коэффициент силуэта**, выбираем k по ним и проверяем соседние значения.

```python
k_range = range(2, 11)
inertias, silhouettes = [], []

for k in k_range:
    km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
    labels = km.fit_predict(X_scaled)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_scaled, labels))

for k, inr, sil in zip(k_range, inertias, silhouettes):
    print(f"k={k}:  inertia={inr:8.1f}   silhouette={sil:.3f}")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
ax[0].plot(list(k_range), inertias, "o-")
ax[0].set_title("Метод локтя (inertia)")
ax[0].set_xlabel("k"); ax[0].set_ylabel("inertia")

ax[1].plot(list(k_range), silhouettes, "o-", color="green")
ax[1].set_title("Коэффициент силуэта")
ax[1].set_xlabel("k"); ax[1].set_ylabel("silhouette")
plt.show()
```

**Выбор k:** локоть заметно сгибается на **k=3**, и силуэт на k=3 один из самых высоких. Соседние k=2 и k=4 проверили: k=2 слишком грубо (только «бедные/богатые»), k=4 уже дробит без явного прироста качества. Берём **k=3** — три понятных уровня развития.

```python
K = 3
kmeans = KMeans(n_clusters=K, random_state=RANDOM_STATE, n_init=10)
clusters = kmeans.fit_predict(X_scaled)
print("Размеры кластеров:")
print(pd.Series(clusters).value_counts().sort_index())
```

## Шаг 3. Создание прогнозного столбца

Номера кластеров от KMeans — это наша новая метка. Добавляем её столбцом и даём кластерам **человеческие имена** по средним признакам, а не «кластер 0/1/2».

```python
df["cluster"] = clusters

# Средние признаки по кластерам (в исходных единицах) — по ним и называем
profile = df.groupby("cluster")[features].mean().round(1)
profile
```

```python
# Назовём кластеры по уровню дохода/ВВП и детской смертности.
# Ранжируем кластеры по gdpp: меньше -> беднее.
order = profile["gdpp"].sort_values().index.tolist()
names = {
    order[0]: "Нужна помощь (низкий доход, высокая детская смертность)",
    order[1]: "Развивающиеся (средний уровень)",
    order[2]: "Развитые (высокий доход, низкая смертность)",
}
df["development_level"] = df["cluster"].map(names)

print("Соответствие кластер -> имя:")
for c in order:
    print(f"  кластер {c}: {names[c]}")

df["development_level"].value_counts()
```

```python
# Несколько стран из каждой группы для проверки осмысленности
for c in order:
    sample = df.loc[df["cluster"] == c, "country"].head(8).tolist()
    print(f"\n{names[c]}:\n  {', '.join(sample)}")
```

## Шаг 4. PCA для проверки

Признаков много — глазом не увидеть. Сжимаем масштабированные данные до 2 главных компонент и раскрашиваем точки по метке из шага 3. Смотрим, отделяются ли кластеры, и **какую долю дисперсии** удержали 2 компоненты.

```python
pca = PCA(n_components=2, random_state=RANDOM_STATE)
coords = pca.fit_transform(X_scaled)
evr = pca.explained_variance_ratio_
print(f"Доля дисперсии: PC1={evr[0]:.1%}, PC2={evr[1]:.1%}, вместе={evr.sum():.1%}")
```

```python
plt.figure(figsize=(9, 6))
for c in order:
    mask = df["cluster"] == c
    plt.scatter(coords[mask, 0], coords[mask, 1], label=names[c], alpha=0.7)
plt.xlabel(f"PC1 ({evr[0]:.1%})")
plt.ylabel(f"PC2 ({evr[1]:.1%})")
plt.title(f"PCA-проекция, раскраска по кластерам (удержано {evr.sum():.1%} дисперсии)")
plt.legend()
plt.show()
```

**Вывод по PCA:** две компоненты удерживают значительную долю дисперсии, и группы на плоскости вытянуты вдоль PC1 (ось «бедные → богатые») и в основном разделимы, хотя на границе средних/развитых есть наложение. Картинка не обманчива, но граница между соседними уровнями размыта — это нормально для непрерывного «уровня развития».

## Шаг 5. Классификация

Проверяем, что метка не случайна: учим классификатор предсказывать нашу метку из шага 3 по исходным признакам. Делим на train/test, считаем accuracy, precision, recall, F1.

```python
y = df["cluster"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y)

# Масштабируем по train, применяем к test (без утечки)
sc = StandardScaler().fit(X_train)
X_train_s = sc.transform(X_train)
X_test_s = sc.transform(X_test)

clf = LogisticRegression(max_iter=1000, random_state=RANDOM_STATE)
clf.fit(X_train_s, y_train)

pred_train = clf.predict(X_train_s)
pred_test = clf.predict(X_test_s)
```

```python
def report(y_true, y_pred, title):
    print(f"=== {title} ===")
    print(f"accuracy : {accuracy_score(y_true, y_pred):.3f}")
    print(f"precision: {precision_score(y_true, y_pred, average='macro'):.3f}")
    print(f"recall   : {recall_score(y_true, y_pred, average='macro'):.3f}")
    print(f"f1       : {f1_score(y_true, y_pred, average='macro'):.3f}\n")

report(y_train, pred_train, "TRAIN")
report(y_test, pred_test, "TEST")
```

```python
print(classification_report(y_test, pred_test))

cm = confusion_matrix(y_test, pred_test)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Матрица ошибок (test)")
plt.xlabel("Предсказано"); plt.ylabel("Истинно")
plt.show()
```

## Вывод команды

- **Данные и находка EDA:** 167 стран, 9 признаков в очень разных масштабах → обязательное масштабирование. Удивило, что детская смертность, рождаемость, продолжительность жизни и доход образуют одну сильную ось — это «уровень развития».
- **Кластеры:** по локтю и силуэту выбрали **k=3**. Назвали их по смыслу: «Нужна помощь», «Развивающиеся», «Развитые» (по доходу/ВВП и детской смертности), а не номерами.
- **PCA:** 2 компоненты удержали большую долю дисперсии; кластеры разделимы вдоль главной оси, граница средних/развитых слегка размыта.
- **Классификация:** логистическая регрессия предсказывает нашу метку с высокими accuracy/F1 и на train, и на test — значит границы между группами реальны и разделимы.

**Вердикт:** структура данных подтвердилась — группы осмысленные (уровни развития стран), а не «разноцветные точки». Гуманитарному фонду помощь стоит направлять в кластер **«Нужна помощь»**.
